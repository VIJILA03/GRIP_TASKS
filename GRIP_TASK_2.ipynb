```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_set="http://bit.ly/w-data"
data=pd.read_csv(data_set)
x=data.iloc[: ,:1].values
y=data.iloc[: ,-1].values
```

```python
data.plot(x='Hours',y='Scores',style="g.")
plt.title("INITIAL CHART")
plt.xlabel("HOURS STUDIED")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
#data preparation
x=data[['Hours']].values
y=data[['Scores']].values
```

```python
#importing the required function and splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#training phase
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(x_train,y_train)
```

```python
inter=lin.intercept_
coeff=lin.coef_
print("The intercept is: ",inter)
print("the cooefficient is:",coeff[0])
```

```python
#plotting regression line
line=lin.coef_*x+lin.intercept_
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
#prediction making
y_pred=lin.predict(x_test)
y_pred
```

```python
y_test
```

```python
#trained data
plt.scatter(x_train,y_train,label="Score")
plt.plot(x_train,lin.predict(x_train),color='red',label='line of best fit')
plt.xlabel("HOURS STUDY")
plt.ylabel("PERCENTAGE")
plt.title("BEST FIT LINE ON TRAINING DATA")
plt.legend()
plt.show()
```

```python
#testing data
plt.scatter(x_test,y_test,label="Score")
plt.plot(x_test,lin.predict(x_test),color='red',label='Line of best fit')
plt.xlabel("HOURS STUDY")
plt.ylabel("PERCENTAGE")
plt.title("BEST FIT LINE ON TESTING DATA")
plt.legend()
plt.show()
```
